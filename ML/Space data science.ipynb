```python
import pandas as pd
import numpy as np
import sklearn as skl
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
```

```python
data = pd.read_csv("problem_dataset.csv")
```

```python
data
```

```python
data.columns
```

```python
delete = ["kepid","tce_plnt_num","tce_rogue_flag","tce_insol","tce_insol_err"]
data_new = data.drop(delete,axis=1)
```

```python
data_new
```

```python
Y = data_new.iloc[:,-1]
X = data_new.iloc[:,:-1]
# train,test = train_test_split

train_X, test_X, train_Y, test_Y = train_test_split(X,Y,test_size = 0.2)
print(train_X.shape)
```

```python
model = RFC(n_estimators=10000)
model.fit(train_X,train_Y)
```

```python
model.score(train_X,train_Y)
```

```python
pred_train = model.predict(train_X)
(np.where(pred_train== train_Y))
```

```python
model.score(test_X,test_Y)
```

```python
pred_test = model.predict(test_X)

len(np.where(pred_test == test_Y)[0])
```

```python
nearest = KNeighborsClassifier(n_neighbors=100)
nearest.fit(train_X,train_Y)
```

```python
nearest.score(train_X,train_Y)
```

```python
normal = StandardScaler()
train_X = normal.fit_transform(train_X)
test_X = normal.fit_transform(test_X)
```

```python
logistic = LogisticRegression(max_iter= 1000000000,tol=1e-20)
logistic.fit(train_X,train_Y)

```

```python
logistic.score(train_X,train_Y)
```

```python
logistic.score(test_X,test_Y)
```

```python
neural = MLPClassifier(hidden_layer_sizes=(640,1280,2560,1280,640,64))
neural.fit(train_X,train_Y)
```

```python
neural.score(train_X,train_Y)
```

```python
neural.score(test_X,test_Y)
```
